```python
df.head()
df
```

```python
Categorical_features=[col for col in df.columns if df[col].dtype=='O']
print(Categorical_features)
# ans=df["SIP Min. Inv.(Rs.)"].sum()
    
df1=df.copy()

df1["SIP Min. Inv.(Rs.)"]=df1["SIP Min. Inv.(Rs.)"].apply(handlevalue)
#importabt to convert categorical varible to numerical

df1["SIP Min. Inv.(Rs.)"] = df1["SIP Min. Inv.(Rs.)"].apply(lambda x: int(float(x.replace(',', ''))))


df1["SIP Min. Inv.(Rs.)"]


Minsip_amount=df1.groupby("Category")["SIP Min. Inv.(Rs.)"].mean().sort_values( ascending=True)

Minsip_amount
plt.figure(figsize=(15, 5))
plt.bar(Minsip_amount.index,Minsip_amount.values,alpha=0.8,color="green")
plt.xticks(rotation=45)


plt.xlabel("Types of Mutual Funds",size=20)
plt.ylabel("Average MIN SIP  ",size=20)
```

# Conversion of categorical column to numerical data  for these fields:
1. Net_Asset_Value(Rs.)
2. CAGR% 6 Months
3. CAGR% 1 Year
4. CAGR% 3 Year
5. 	Min. Invest(Rs.)
6. 	Exp. Ratio(%)


```python
def handlevalue(value):
    value=str(value).split('.')
    value=value[0]
    return (value)
# For ["Net_Asset_Value(Rs.)"]

df1["Net_Asset_Value(Rs.)"]=df1["Net_Asset_Value(Rs.)"].apply(handlevalue)
df1["Net_Asset_Value(Rs.)"] = df1["Net_Asset_Value(Rs.)"].apply(lambda x: int(float(x.replace(',', ''))))
df1["Net_Asset_Value(Rs.)"]

# For CAGR% 6 Months
rows_to_drop = df1[df1['CAGR% 6 Months'] == '-'].index
df1=df1.drop(rows_to_drop)


df1["CAGR% 6 Months"] = df1["CAGR% 6 Months"].apply(lambda x: int(float(x.replace('.', ''))))
df1["CAGR% 6 Months"]
df1["CAGR% 6 Months"]=(df1["CAGR% 6 Months"]/100)
df1["CAGR% 6 Months"]

#For CAGR% 1 Year
rows_to_drop = df1[df1['CAGR% 1 Year'] == '-'].index
df1=df1.drop(rows_to_drop)
df1["CAGR% 1 Year"] = df1["CAGR% 1 Year"].apply(lambda x: int(float(x.replace('.', ''))))
df1["CAGR% 1 Year"]
df1["CAGR% 1 Year"]=(df1["CAGR% 1 Year"]/100)
df1["CAGR% 1 Year"]

# For 'CAGR% 3 Year'
rows_to_drop = df1[df1['CAGR% 3 Year'] == '-'].index
df1=df1.drop(rows_to_drop)
df1['CAGR% 3 Year'] = df1['CAGR% 3 Year'].apply(lambda x: int(float(x.replace('.', ''))))
df1['CAGR% 3 Year']
df1['CAGR% 3 Year']=(df1['CAGR% 3 Year']/100)
df1['CAGR% 3 Year']

#For Min. Invest(Rs.)

df1["Min. Invest(Rs.)"]=df1["Min. Invest(Rs.)"].apply(handlevalue)


df1["Min. Invest(Rs.)"] = df1["Min. Invest(Rs.)"].apply(lambda x: int(float(x.replace(',', ''))))
df1['Min. Invest(Rs.)']

#For Exp. Ratio(%)
df1["Exp. Ratio(%)"] = df1["Exp. Ratio(%)"].apply(lambda x: int(float(x.replace('.', ''))))
df1["Exp. Ratio(%)"]=df1["Exp. Ratio(%)"]/100
df1["Exp. Ratio(%)"]


```
